```python
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
def unzip(l):
    return list(zip(*l))

def sample(ancestral, p=0.1):
        ind = np.copy(ancestral)
        for i in range(len(ind)):
            if np.random.random() < p:
                #ind[i] = np.random.choice(["A","C","G","T"])
                ind[i] = np.random.choice(list(filter(lambda x: x!=ancestral[i],["A","C","G","T"])))
        return ind
```

```python
def make_sfs(num_loci,num_samples,p):
    ancestral = np.random.choice(["A","C","G","T"], size=num_loci)

    pop = np.array([sample(ancestral,p=p) for i in range(num_samples)]).T
    diffs = [(s != a).sum() for s,a in zip(pop,ancestral)]

    sfs = defaultdict(lambda:0)
    for diff in diffs:
        sfs[diff]+=1
    sfs = sorted(list(sfs.items()))
    return sfs
    
```

```python
sfs = make_sfs(num_loci=200,num_samples=1000,p=.001)
daf, sites = unzip(sfs)
daf,sites
```
